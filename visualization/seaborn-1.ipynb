```python
df2 =sb.load_dataset('tips')
sb.jointplot(x='total_bill',y='tip',data=df2);
plt.show()
```

```python
sb.set_style("ticks")
sb.pairplot(df2,hue='sex',diag_kind="kde",kind="scatter")
plt.show()
```

```python
sb.stripplot(x="sex", y="total_bill", data=df2)
plt.show()
```

```python
sb.stripplot(x="day", y="total_bill", data=df2)
plt.show()
```

```python
sb.stripplot(x="time", y="total_bill", data=df2)
plt.show()
```

```python
sb.stripplot(x="sex", y="tip", data=df2, jitter=True)
plt.show()
```

```python
sb.swarmplot(x="sex", y="tip", data=df2)
plt.show()
```
