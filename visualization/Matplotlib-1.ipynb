```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,3,2,4])
plt.show()
```

```python
x = range(6)
plt.plot(x, [xi**2 for xi in x]) 
plt.show()
```

```python
x = range(1, 5) 
plt.plot(x, [xi*1.5 for xi in x],label='Normal') 
plt.plot(x, [xi*3.0 for xi in x],label='Fast') 
plt.plot(x, [xi/3.0 for xi in x],label='Slow') 
plt.xlabel('This is the X axis')
plt.ylabel('This is the Y axis')
plt.legend()
plt.grid()
plt.title('Simple plot')
plt.show()
```

```python
y = np.arange(1, 3) 
plt.plot(y, 'y');    #yellow
plt.plot(y+1, 'm');  #magenta
plt.plot(y+2, 'c'); #cyan
plt.show()
```

```python
y = np.arange(1, 3)
plt.plot(y, '--', y+1, '-.', y+2, ':'); 
plt.show()
```

```python
y = np.arange(1, 3, 0.2)
plt.plot(y, 'x', y+0.5, 'o', y+1, 'D', y+1.5, '^', y+2, 's');
plt.show()
```

```python
y = np.arange(1, 3, 0.3)
plt.plot(y, color='blue', linestyle='dashdot', linewidth=4, marker='o', markerfacecolor='red', markeredgecolor='black', markeredgewidth=3, markersize=12);
plt.show()
```

# Histograms

```python
y= np.random.randn(1000)  #By default, hist() uses a bin value of 10 (so only ten categories, or bars, are computed)
plt.hist(y)
plt.show()
```

```python
plt.hist(y, 25)  #hist(y, bin=<bins>)
plt.show()
```

# Error Bar Charts

```python
x = np.arange(0, 4, 0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))  #can also declare e2 and then xerr=e2
plt.errorbar(x, y, yerr=e1, fmt='.-');
plt.show()
```

# Bar Charts

```python
plt.bar([1, 2, 3], [3, 2, 5]);
plt.show()
```

```python
dict = {'A': 40, 'B': 70, 'C': 30, 'D': 85} 
plt.bar(range(len(dict)), list(dict.values()));
plt.xticks(range(len(dict)), list(dict.keys())) # CBDA below the bars
plt.show() 
```

```python
data1 = 10*np.random.rand(5)
data2 = 10*np.random.rand(5)
data3 = 10*np.random.rand(5)
e2 = 0.5 * np.abs(np.random.randn(len(data2)))  # yerr= e2
locs = np.arange(1, len(data1)+1) 
width = 0.27 
plt.bar(locs, data1, width=width);
plt.bar(locs+width, data2, yerr=e2, width=width, color='red'); 
plt.bar(locs+2*width, data3, width=width, color='green') ;
plt.xticks(locs + width*1.5, locs);
plt.show()
```
